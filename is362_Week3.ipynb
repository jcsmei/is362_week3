```python
import numpy.random as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy

%matplotlib inline
```

Function to generate random days and random numbers of miles

```python
def CreateDataSet(Number=1):
    output = []
    for i in range(Number):
        rng = pd.period_range(start='2018-01-01', end='2018-01-30', freq='d')
        data = np.randint(low=0, high=125, size=len(rng))
        output.extend(zip(rng, data))
    return output
```

```python
dataset = CreateDataSet(2)
df = pd.DataFrame(data=dataset, columns=['Day', 'Miles'])
df.info()
```
